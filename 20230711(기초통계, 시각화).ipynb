```python
from statannot import add_stat_annotation
```

```python
f1 = sns.boxplot(data=mpg, x='o', y='horsepower')
f2 = sns.stripplot(data=mpg, x='o', y='horsepower', color='k', alpha=.5)

add_stat_annotation(f1, data=mpg, x='o', y='horsepower',
                    box_pairs=[(('over'),('under'))],
                    test='t-test_ind', text_format='star', loc='inside', verbose=2)
#verbose 검정결과 출력 0:안알려줌 1:일부만 2:전부
```

```python
# user_df: 입력할 데이터프레임 이름
# col: x축 column 이름
# value: y축 value 이름
# group1, group2: col 기준으로 분류된 이름

def ttest_auto(user_df, col, value, group1, group2):

    f1 = sns.boxplot(data=user_df, x=col, y=value)
    f2 = sns.stripplot(data=user_df, x=col, y=value, color='k', alpha=.5)

    add_stat_annotation(f1, data=user_df, x=col, y=value,
                        box_pairs=[((group1),(group2))],
                        test='t-test_ind', text_format='star', loc='inside', verbose=2)
    
    plt.show()
```

* penguins 데이터프레임에서,Adelie과 Gentoo의 체중 차이가 나는지를 ttest_auto 함수를 이용해 표현해보자

```python
penguins['species'].unique()
```

```python
# 결측치 처리
penguins['body_mass_g'] = [penguins['body_mass_g'].median() if w != w else w for w in penguins['body_mass_g']]
```

```python
penguins['body_mass_g'].unique()
```

```python
ttest_auto(penguins, 'species', 'body_mass_g', 'Adelie', 'Gentoo')
#정규성 검정, 등분산성 검정 필요

# if 정규성 만족: ttest_ind 실행
#           불만족: (등분산성 만족) welch's ttest 실행


# 다변량 사후검정 필요
# 여러번 검정할수록 p-value(오차가 발생할 확률)누적됨 -> 0.05/3=0.0167이 새로운 cutoff
# cutoff 그대로 두고 p-value*3
```

```python
f1 = sns.boxplot(data=mpg, x='o', y='horsepower')
f2 = sns.stripplot(data=mpg, x='o', y='horsepower', color='k', alpha=.5)

add_stat_annotation(f1, data=mpg, x='o', y='horsepower',
                    box_pairs=[(('over'),('under'))],
                    test='t-test_ind', text_format='star', loc='inside', verbose=1)
```
