## 매각 금액이 가장 많은 팀과 하위 팀의 차이 확인하기

```python
df['팀명'].unique()
```

```python
q1 = df.groupby(['부점명','팀명']).agg({'대장금액(원)' : 'sum'}).sort_values(by='대장금액(원)', ascending=False)
q1
```

귀무가설: 매각 금액이 가장 많은 팀과 하위 팀의 차이가 없을 것이다.

```python
max_sale = (df['부점명']=='서울서부지역본부') & (df['팀명']=='국유재산3팀')
```

```python
from scipy import stats
```

```python
d1 = df[max_sale]['대장금액(원)']
d2 = df[-max_sale]['대장금액(원)']
```

```python
stats.shapiro(d1)
# 정규성 만족하지 않는다.
```

```python
stats.shapiro(d2)
# 정규성 만족하지 않는다.
```

샘플 개수가 충분히 많기 때문에 정규성 만족한다고 볼 수 있다.

```python
#등분산성 검정
stats.ks_2samp(d1, d2)
#같은 집단에서 추출했으므로 모집단 분산 같음. 등분산성 만족한다고 가정 가능
```

```python
# 정규성, 등분산성 만족
stats.ttest_ind(d1, d2)
```

```python
#정규성 불만족
stats.mannwhitneyu(d1, d2)
```

pvalue가 0.05미만으로 귀무가설 기각
